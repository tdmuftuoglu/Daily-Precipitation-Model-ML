```python
# ==============================================================================
# NEWSTUDY – ONE-BLOCK COLAB PIPELINE (Bilingual / Çift dilli)
# ==============================================================================
# Author / Yazar: Dr. Tevfik Denizhan MUFTUOGLU
#
# License & Attribution / Lisans ve Atıf:
# - This code was developed by Dr. Tevfik Denizhan MUFTUOGLU.
# - It may be used and/or modified **with proper citation**.
# - Bu kod Dr. Tevfik Denizhan MUFTUOGLU tarafından geliştirilmiştir.
# - **Atıf verilmesi** kaydıyla kullanılabilir ve/veya değiştirilebilir.
#
# Recommended citation / Önerilen atıf:
# MUFTUOGLU, T. D. (2025). NEWSTUDY: Hybrid and ML-based daily precipitation
# forecasting with significance tests, return levels, and explainability.
# (Code & results). DOI/URL: <your link>.
#
# ------------------------------- WHAT THIS DOES --------------------------------
# EN: End-to-end reproducible pipeline:
#     - Installs deps, loads NASA-POWER CSV, builds datetime index
#     - Independent stats: (i) Gumbel return levels + 95% CI (bootstrap)
#                          (ii) Mann–Kendall test + centered Theil–Sen line
#     - Feature engineering: proper lag creation (1..3) for target & predictors
#     - Model suite (leakage-safe): XGB, LGBM, RF, Prophet, SARIMAX, ANN,
#                                   LSTM, GRU, Hybrid Prophet+XGB (***FIXED***)
#     - Baselines: Persistence (t-1), Seasonal Naive (t-365d)
#     - Event-based verification: P90 & 20 mm/day (HitRate, FAR, CSI)
#     - SHAP explainability (XGB/LGBM/RF + optional ANN)
#     - Architecture diagrams (RF/ANN/Stacking) + ZIP download
#
# TR: Baştan sona tekrarlanabilir boru hattı:
#     - Kütüphaneler kurulur, NASA-POWER CSV yüklenir, tarih indeksi kurulur
#     - Bağımsız analizler: (i) Gumbel tasarım yağışları + %95 GA (bootstrap)
#                           (ii) Mann–Kendall testi + merkezlenmiş Theil–Sen
#     - Özellik üretimi: hedef ve girdiler için 1..3 gecikme (lag)
#     - Model seti (sızıntısız): XGB, LGBM, RF, Prophet, SARIMAX, ANN, LSTM,
#                                GRU, Hibrid Prophet+XGB (***DÜZELTİLDİ***)
#     - Basit karşılaştırmalar: Süreklilik (t-1), Mevsimsel-naif (t-365)
#     - Olay-tabanlı doğrulama: P90 ve 20 mm/gün (Vuruş, Yanlış Alarm, CSI)
#     - Açıklanabilirlik: SHAP (XGB/LGBM/RF + isteğe bağlı ANN)
#     - Mimari şemalar (RF/ANN/Stacking) + ZIP indirme
# ==============================================================================

# -------------------------- 0) Install & Imports / Kurulum --------------------
print(">>> Installing libraries / Kütüphaneler yükleniyor...")
!pip -q install pymannkendall prophet xgboost lightgbm shap graphviz pydot

from google.colab import files
import os, random, io, zipfile, warnings, math
warnings.filterwarnings("ignore")

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
plt.style.use('seaborn-v0_8-whitegrid')

import pymannkendall as mk
from scipy.stats import gumbel_r

from sklearn.model_selection import TimeSeriesSplit, RandomizedSearchCV
from sklearn.preprocessing import StandardScaler, MinMaxScaler
from sklearn.metrics import mean_squared_error, r2_score
from sklearn.ensemble import RandomForestRegressor, StackingRegressor
from sklearn.linear_model import LinearRegression

import xgboost as xgb
import lightgbm as lgb
from prophet import Prophet
from statsmodels.tsa.statespace.sarimax import SARIMAX

import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import LSTM, GRU, Dense

import shap
from graphviz import Digraph

# Helper for bilingual step headers / Çift dilli adım başlıkları
def step(en_title, tr_title=None):
    bar = "="*78
    print(f"\n{bar}\n[STEP] {en_title}")
    if tr_title:
        print(f"[ADIM] {tr_title}")
    print(bar)

# -------------------------- 1) Reproducibility / Tekrarlanabilirlik -----------
SEED = 42
os.environ['PYTHONHASHSEED'] = str(SEED)
random.seed(SEED); np.random.seed(SEED); tf.random.set_seed(SEED)
print(">>> Seeds set to 42 / Tohumlar 42 olarak ayarlandı.")

# -------------------------- 2) Data Upload / Veri Yükleme ---------------------
step("Upload the NASAPOWERDATASET.csv file",
     "NASAPOWERDATASET.csv dosyasını yükleyin")
print("Use the dialog to select your CSV / Dosya seçim iletişim kutusunu kullanın")
uploaded = files.upload()
csv_name = next(iter(uploaded))
df = pd.read_csv(csv_name)

# Build datetime index from YEAR + DOY / YEAR+DOY'dan zaman indeksi kur
df['date'] = pd.to_datetime(df['YEAR'].astype(str) + '-' + df['DOY'].astype(str),
                            format='%Y-%j')
df = df.sort_values('date').set_index('date')

TARGET = 'PRECTOTCORR'
POTENTIAL_FEATURES = ['T2M','RH2M','PS','WS10M','WD10M','ALLSKY_SFC_SW_DWN']

# -------------------------- 3) Independent Stats / Bağımsız Analizler --------
# 3.1) Return Periods (Gumbel + 95% CI)
step("Return period analysis (Gumbel) with 95% bootstrap CI",
     "Tekerrür analizi (Gumbel) ve %95 bootstrap güven aralığı")
annual_max = df[TARGET].resample('Y').max().dropna()
loc, scale = gumbel_r.fit(annual_max.values)
RPs = np.array([2,5,10,25,50,100])
rl = gumbel_r.ppf(1 - 1.0/RPs, loc=loc, scale=scale)

BOOT = 300  # ↑ for tighter CIs / aralıkları daraltmak için arttırılabilir
rng = np.random.default_rng(SEED)
rl_boot = np.zeros((BOOT, len(RPs)))
vals = annual_max.values
for b in range(BOOT):
    sample = rng.choice(vals, size=len(vals), replace=True)
    l_b, s_b = gumbel_r.fit(sample)
    rl_boot[b,:] = gumbel_r.ppf(1 - 1.0/RPs, loc=l_b, scale=s_b)

rl_lo = np.percentile(rl_boot, 2.5, axis=0)
rl_hi = np.percentile(rl_boot,97.5, axis=0)

gumbel_table = pd.DataFrame({
    'ReturnPeriod_years': RPs,
    'ReturnLevel_mm_day': np.round(rl,2),
    'CI95_low': np.round(rl_lo,2),
    'CI95_high': np.round(rl_hi,2)
})
gumbel_table.to_csv('design_rainfall_table_with_CI.csv', index=False)

plt.figure(figsize=(11,7))
plt.plot(RPs, rl, marker='o', label='Return level (Gumbel MLE)')
plt.fill_between(RPs, rl_lo, rl_hi, alpha=0.25, label='95% CI (bootstrap)')
plt.xscale('log'); plt.xticks(RPs, RPs)
plt.xlabel('Return period (years)'); plt.ylabel('Maximum daily precipitation (mm/day)')
plt.title('Return Period vs Return Level (Gumbel, with 95% CI)')
plt.legend(); plt.tight_layout()
plt.savefig('GUMBEL_return_levels_CI.png', dpi=300)

# 3.2) Trend: Mann–Kendall + centered Theil–Sen / Merkezli Theil–Sen
step("Trend of annual maxima (Mann–Kendall + centered Theil–Sen)",
     "Yıllık maksimumların eğilimi (Mann–Kendall + merkezli Theil–Sen)")
mk_series = annual_max.copy()
mk_res = mk.original_test(mk_series)
years = mk_series.index.year.values.astype(float)
x0 = np.median(years)
trend_line = np.median(mk_series.values) + mk_res.slope * (years - x0)

plt.figure(figsize=(14,8))
plt.plot(years, mk_series.values, '-o', label='Annual Maximum Daily Precipitation')
plt.plot(years, trend_line, 'r--', label=f'Theil–Sen Trend (slope={mk_res.slope:.3f}, p={mk_res.p:.3f})')
plt.title('Trend in Annual Maximum Daily Precipitation')
plt.xlabel('Year'); plt.ylabel('Maximum Daily Precipitation (mm/day)')
plt.legend(); plt.grid(True); plt.tight_layout()
plt.savefig('FINAL_TREND_ANALYSIS_fixed.png', dpi=300)

# -------------------------- 4) Feature Engineering / Özellik Üretimi ----------
step("Create 1..3 lags for target & all predictors",
     "Hedef ve tüm girdiler için 1..3 gecikme (lag) üret")
df_model = df[[TARGET] + POTENTIAL_FEATURES].copy()
for col in [TARGET] + POTENTIAL_FEATURES:
    for lag in (1,2,3):
        df_model[f'{col}_lag_{lag}'] = df_model[col].shift(lag)
df_model = df_model.dropna().copy()

# Train/test split by time / Zaman bazlı eğitim/test
split_idx = int(len(df_model)*0.80)
train_df = df_model.iloc[:split_idx].copy()
test_df  = df_model.iloc[split_idx:].copy()
y_train = train_df[TARGET].copy()
y_test  = test_df[TARGET].copy()
X_cols_all = [c for c in df_model.columns if c != TARGET]

# Quick importance to pick TOP_N / Hızlı önem grafiği ile TOP_N seçim
step("Select top features via XGBoost importance",
     "XGBoost önem puanlarıyla en iyi öznitelikleri seç")
xgb_explorer = xgb.XGBRegressor(
    n_estimators=300, random_state=SEED, n_jobs=-1, objective='reg:squarederror'
)
xgb_explorer.fit(train_df[X_cols_all], y_train, verbose=False)
imp_df = pd.DataFrame({'feature': X_cols_all,
                       'importance': xgb_explorer.feature_importances_}
                      ).sort_values('importance', ascending=False)
TOP_N = 15
TOP_FEATURES = imp_df.head(TOP_N)['feature'].tolist()

plt.figure(figsize=(11,8))
sns.barplot(x='importance', y='feature', data=imp_df.head(TOP_N))
plt.title('Feature Importance Ranking (XGBoost)'); plt.xlabel('Importance'); plt.ylabel('Feature')
plt.tight_layout(); plt.savefig('feature_importance_FIXED.png', dpi=300)

# -------------------------- 5) Models & Baselines / Modeller ------------------
step("Train baselines and ML/TS models (leakage-safe)",
     "Basit modeller ve ML/Zaman serisi modellerini eğit (sızıntısız)")

def rmse(y_true, y_pred): return math.sqrt(mean_squared_error(y_true, y_pred))
performance, pred_store = [], {}

def log_model(name, y_true_index, preds):
    s = pd.Series(preds, index=y_true_index)
    common = y_test.index.intersection(s.index)
    s = s.loc[common]
    r2v = r2_score(y_test.loc[common], s.values)
    rmsev = rmse(y_test.loc[common], s.values)
    performance.append({'Model': name, 'RMSE (mm/day)': rmsev, 'R-squared (R2)': r2v})
    pred_store[name] = s

# 5.1) Baselines / Basit modeller
pers = test_df[f'{TARGET}_lag_1'];      log_model('Baseline - Persistence (t-1)', pers.index, pers.values)
daily = df[[TARGET]].asfreq('D').interpolate()
seasonal_pred = daily[TARGET].shift(365).reindex(y_test.index).interpolate()
log_model('Baseline - Seasonal Naive (t-365d)', seasonal_pred.index, seasonal_pred.values)

# 5.2) Trees / Ağaçlar
X_tr, X_te = train_df[TOP_FEATURES], test_df[TOP_FEATURES]
model_xgb = xgb.XGBRegressor(n_estimators=800, learning_rate=0.05,
                             max_depth=6, subsample=0.9, colsample_bytree=0.9,
                             random_state=SEED, objective='reg:squarederror')
model_xgb.fit(X_tr, y_train, eval_set=[(X_te, y_test)], verbose=False)
log_model('XGBoost', y_test.index, model_xgb.predict(X_te))

model_lgb = lgb.LGBMRegressor(n_estimators=800, learning_rate=0.05,
                              subsample=0.9, colsample_bytree=0.9,
                              random_state=SEED)
model_lgb.fit(X_tr, y_train, eval_set=[(X_te, y_test)],
              callbacks=[lgb.early_stopping(50, verbose=False)])
log_model('LightGBM', y_test.index, model_lgb.predict(X_te))

model_rf = RandomForestRegressor(n_estimators=300, random_state=SEED, n_jobs=-1)
model_rf.fit(X_tr, y_train)
log_model('Random Forest', y_test.index, model_rf.predict(X_te))

# 5.3) Prophet (exogenous without lags) / Prophet (dışsal değişkenli, lagsiz)
feat_for_prophet = [c for c in POTENTIAL_FEATURES if c in df.columns]
prop_df = df[[TARGET] + feat_for_prophet].reset_index().rename(columns={'date':'ds', TARGET:'y'}).dropna()
split_prop = int(len(prop_df)*0.80)
train_prop, test_prop = prop_df.iloc[:split_prop], prop_df.iloc[split_prop:]
m_prophet = Prophet()
for c in feat_for_prophet: m_prophet.add_regressor(c)
m_prophet.fit(train_prop)
pred_prop = m_prophet.predict(test_prop.drop(columns=['y']))['yhat'].values
log_model('Prophet', test_prop['ds'], pred_prop)

# 5.4) SARIMAX (with exog) / Dışsal değişkenli SARIMAX
exog_tr, exog_te = train_df[POTENTIAL_FEATURES], test_df[POTENTIAL_FEATURES]
model_sarimax = SARIMAX(endog=y_train, exog=exog_tr, order=(1,1,1), seasonal_order=(0,0,0,0)).fit(disp=False)
log_model('SARIMAX', y_test.index, model_sarimax.get_forecast(steps=len(y_test), exog=exog_te).predicted_mean.values)

# 5.5) ANN (scaling on train only) / ANN (yalnızca train ölçekleme)
sc_ann = StandardScaler().fit(X_tr.values)
Xtr_ann, Xte_ann = sc_ann.transform(X_tr.values), sc_ann.transform(X_te.values)
ann = Sequential([Dense(64, activation='relu', input_shape=(Xtr_ann.shape[1],)),
                  Dense(32, activation='relu'),
                  Dense(1)])
ann.compile(optimizer='adam', loss='mse')
ann.fit(Xtr_ann, y_train.values, epochs=50, batch_size=32, verbose=0)
log_model('ANN', y_test.index, ann.predict(Xte_ann, verbose=0).flatten())

# 5.6) LSTM & GRU (leakage-safe MinMax) / sızıntısız MinMax
DL_FEATURES = [TARGET] + POTENTIAL_FEATURES
sc_dl = MinMaxScaler().fit(train_df[DL_FEATURES].values)
train_scaled = sc_dl.transform(train_df[DL_FEATURES].values)
test_scaled  = sc_dl.transform(test_df[DL_FEATURES].values)

def to_seq(data, time_step=30):
    X, y = [], []
    for i in range(len(data)-time_step-1):
        X.append(data[i:i+time_step, :])
        y.append(data[i+time_step, 0])  # first col is target
    return np.array(X), np.array(y)

time_step = 30
Xtr_dl, ytr_dl = to_seq(train_scaled, time_step)
Xte_dl, yte_dl = to_seq(test_scaled, time_step)

lstm = Sequential([LSTM(50, input_shape=(Xtr_dl.shape[1], Xtr_dl.shape[2])), Dense(1)])
lstm.compile(optimizer='adam', loss='mse')
lstm.fit(Xtr_dl, ytr_dl, epochs=20, batch_size=64, verbose=0)
pred_lstm_sc = lstm.predict(Xte_dl, verbose=0).flatten()
dummy = np.zeros((len(pred_lstm_sc), len(DL_FEATURES))); dummy[:,0] = pred_lstm_sc
pred_lstm = sc_dl.inverse_transform(dummy)[:,0]
y_test_dl = y_test.iloc[time_step+1:]
log_model('LSTM', y_test_dl.index, pred_lstm)

gru = Sequential([GRU(50, input_shape=(Xtr_dl.shape[1], Xtr_dl.shape[2])), Dense(1)])
gru.compile(optimizer='adam', loss='mse')
gru.fit(Xtr_dl, ytr_dl, epochs=20, batch_size=64, verbose=0)
pred_gru_sc = gru.predict(Xte_dl, verbose=0).flatten()
dummy2 = np.zeros((len(pred_gru_sc), len(DL_FEATURES))); dummy2[:,0] = pred_gru_sc
pred_gru = sc_dl.inverse_transform(dummy2)[:,0]
log_model('GRU', y_test_dl.index, pred_gru)

# 5.7) Hybrid Prophet+XGB on residuals (***LEAKAGE FIXED***)
step("Hybrid (Prophet + XGB on residuals) with TimeSeriesSplit (LEAKAGE FIXED)",
     "Hibrid (Prophet + artıklara XGB) – TimeSeriesSplit ile (SIZINTI DÜZELTİLDİ)")
df_h = df[[TARGET]].reset_index().rename(columns={'date':'ds', TARGET:'y'}).dropna()
split_h = int(len(df_h)*0.80)
train_h = df_h.iloc[:split_h]
prop_base = Prophet().fit(train_h[['ds','y']])
forecast_full = prop_base.predict(df_h[['ds']])['yhat'].values
df_h['prophet_forecast'] = forecast_full
df_h['residuals'] = df_h['y'] - df_h['prophet_forecast']

# Lag frame (no current target) / Gecikmeli çerçeve
df_lags = df[[TARGET] + POTENTIAL_FEATURES].copy()
for col in [TARGET] + POTENTIAL_FEATURES:
    for lag in (1,2,3):
        df_lags[f'{col}_lag_{lag}'] = df_lags[col].shift(lag)
df_lags = df_lags.dropna().reset_index().rename(columns={'date':'ds'})

# Merge with prophet outputs and set index to 'ds'
df_h2 = pd.merge(df_lags, df_h[['ds','y','prophet_forecast','residuals']],
                 on='ds', how='inner').set_index('ds')

Y_h = df_h2['residuals']
# CRITICAL: exclude residuals, the true target, current TARGET & prophet_forecast
X_h = df_h2.drop(columns=['residuals', 'y', TARGET, 'prophet_forecast'])

split_h2 = int(len(df_h2)*0.80)
Xh_tr, Xh_te = X_h.iloc[:split_h2], X_h.iloc[split_h2:]
Yh_tr, Yh_te = Y_h.iloc[:split_h2], Y_h.iloc[split_h2:]

# >>> FIXED INDEX MATCH: take prophet forecast from df_h2 (same index as Xh_te)
prop_test_forecast = df_h2.loc[Xh_te.index, 'prophet_forecast'].values

tscv = TimeSeriesSplit(n_splits=5)
param_grid = {
    'n_estimators':[200,600,1000],
    'max_depth':[3,5,8],
    'learning_rate':[0.01,0.05,0.1],
    'subsample':[0.7,0.9,1.0],
    'colsample_bytree':[0.7,0.9,1.0]
}
rs = RandomizedSearchCV(
    xgb.XGBRegressor(random_state=SEED, objective='reg:squarederror'),
    param_distributions=param_grid, n_iter=20, scoring='r2',
    cv=tscv, n_jobs=-1, random_state=SEED, verbose=0
)
rs.fit(Xh_tr, Yh_tr)
best_h = rs.best_estimator_
res_pred = best_h.predict(Xh_te)
hyb_pred = prop_test_forecast + res_pred
log_model('Hybrid (Optimized)', Xh_te.index, hyb_pred)

# 5.8) Stacking Ensemble / Yığın model
estimators = [
    ('xgb', xgb.XGBRegressor(n_estimators=300, random_state=SEED, objective='reg:squarederror')),
    ('lgbm', lgb.LGBMRegressor(n_estimators=300, random_state=SEED)),
    ('rf', RandomForestRegressor(n_estimators=200, random_state=SEED, n_jobs=-1))
]
stacking = StackingRegressor(estimators=estimators, final_estimator=LinearRegression(), n_jobs=-1)
stacking.fit(X_tr, y_train)
log_model('Stacking Ensemble', y_test.index, stacking.predict(X_te))

# -------------------------- 6) Results & Plots / Sonuçlar ---------------------
step("Generate tables & plots", "Tabloları ve grafikleri üret")

perf_df = pd.DataFrame(performance).sort_values('R-squared (R2)', ascending=False)
perf_df.to_csv('ULTIMATE_FINAL_comparison_FIXED.csv', index=False)
print("\n=== Final Model Comparison (sorted by R2) ===")
print(perf_df)

plt.figure(figsize=(13,9))
sns.barplot(x='R-squared (R2)', y='Model', data=perf_df, palette='viridis')
plt.title('Final Model Performance Comparison (R-squared)')
plt.xlabel('R-squared (R2)'); plt.ylabel('Model')
plt.tight_layout(); plt.savefig('model_performance_barchart_FIXED.png', dpi=300)

top6 = perf_df.head(6)['Model'].tolist()
plt.figure(figsize=(19,9))
plt.plot(y_test.index, y_test.values, color='k', linewidth=2, alpha=0.6, label='Actual')
styles = ['r-','b--','g:','m-.','c:','y--']
for i, m in enumerate(top6):
    s = pred_store[m]
    common = y_test.index.intersection(s.index)
    plt.plot(common, s.loc[common].values, styles[i%len(styles)],
             label=f"{m} (R2={perf_df.set_index('Model').loc[m,'R-squared (R2)']:.3f})", linewidth=1.5)
plt.title('Final Comparison of Top Performing Models')
plt.xlabel('Date'); plt.ylabel('Precipitation (mm/day)'); plt.legend(ncol=2)
plt.tight_layout(); plt.savefig('ULTIMATE_FINAL_comparison_FIXED.png', dpi=300)

champ = perf_df.iloc[0]['Model']
s = pred_store[champ]; common = y_test.index.intersection(s.index)
plt.figure(figsize=(9,9))
plt.scatter(y_test.loc[common].values, s.loc[common].values, alpha=0.5)
mx = max(y_test.loc[common].max(), s.loc[common].max())
plt.plot([0,mx],[0,mx],'r--',label='Perfect prediction')
plt.title(f'Champion Model: {champ} (R2={perf_df.iloc[0]["R-squared (R2)"]:.3f})')
plt.xlabel('Actual (mm/day)'); plt.ylabel('Predicted (mm/day)'); plt.legend(); plt.axis('equal')
plt.tight_layout(); plt.savefig('champion_model_scatter_FIXED.png', dpi=300)

# -------------------------- 7) Event-based Metrics / Olay Metrikleri ----------
step("Event-based verification (P90 & 20 mm/day thresholds)",
     "Olay-tabanlı doğrulama (P90 ve 20 mm/gün eşikleri)")
thr1 = y_test.quantile(0.90); thr2 = 20.0

def event_metrics(y_true, y_pred, thr):
    yt = (y_true >= thr).astype(int); yp = (y_pred >= thr).astype(int)
    TP = int(((yt==1)&(yp==1)).sum()); FP = int(((yt==0)&(yp==1)).sum())
    FN = int(((yt==1)&(yp==0)).sum()); TN = int(((yt==0)&(yp==0)).sum())
    hit_rate = TP / max(1, (TP+FN)); far = FP / max(1, (TP+FP))
    csi = TP / max(1, (TP+FP+FN))
    return TP,FP,FN,TN,hit_rate,far,csi

rows = []
for m, s in pred_store.items():
    common = y_test.index.intersection(s.index)
    y_true_c, y_pred_c = y_test.loc[common], s.loc[common]
    r1, r2 = event_metrics(y_true_c, y_pred_c, thr1), event_metrics(y_true_c, y_pred_c, thr2)
    rows.append({'Model': m,
                 'THR_P90': float(thr1), 'P90_HitRate': round(r1[4],3),
                 'P90_FAR': round(r1[5],3), 'P90_CSI': round(r1[6],3),
                 'THR_20mm': thr2, '20mm_HitRate': round(r2[4],3),
                 '20mm_FAR': round(r2[5],3), '20mm_CSI': round(r2[6],3)})
events_df = pd.DataFrame(rows).sort_values('P90_CSI', ascending=False)
events_df.to_csv('EVENT_METRICS_top_models.csv', index=False)

# -------------------------- 8) SHAP Explainability / SHAP Açıklanabilirlik ----
step("SHAP global explanations for tree models (+optional ANN)",
     "Ağaç modelleri için SHAP global açıklamalar (+isteğe bağlı ANN)")
Xte_df = test_df[TOP_FEATURES].copy()
SAMPLE_SHAP = min(1000, len(Xte_df))
Xte_sample = Xte_df.iloc[:SAMPLE_SHAP]

def save_tree_shap(model, Xsample, name_prefix):
    expl = shap.TreeExplainer(model)
    sv = expl.shap_values(Xsample)
    plt.figure(); shap.summary_plot(sv, Xsample, plot_type='bar', show=False)
    plt.tight_layout(); plt.savefig(f'{name_prefix}_SHAP_bar.png', dpi=300); plt.close()
    plt.figure(); shap.summary_plot(sv, Xsample, show=False)
    plt.tight_layout(); plt.savefig(f'{name_prefix}_SHAP_beeswarm.png', dpi=300); plt.close()

save_tree_shap(model_rf, Xte_sample, 'RF')
save_tree_shap(model_lgb, Xte_sample, 'LGBM')
save_tree_shap(model_xgb, Xte_sample, 'XGB')

try:
    step("Optional: ANN SHAP (KernelExplainer, sampled)",
         "İsteğe bağlı: ANN SHAP (KernelExplainer, örneklemeli)")
    K_SAMP = min(200, len(Xtr_ann))
    bkg = Xtr_ann[:K_SAMP]
    Xtest_ann_sample = Xte_ann[:min(200, len(Xte_ann))]
    def ann_pred(X): return ann.predict(X, verbose=0).flatten()
    kexpl = shap.KernelExplainer(ann_pred, bkg)
    ksv = kexpl.shap_values(Xtest_ann_sample, nsamples=100)
    plt.figure(); shap.summary_plot(ksv, Xtest_ann_sample, feature_names=TOP_FEATURES, plot_type='bar', show=False)
    plt.tight_layout(); plt.savefig('ANN_SHAP_bar.png', dpi=300); plt.close()
    plt.figure(); shap.summary_plot(ksv, Xtest_ann_sample, feature_names=TOP_FEATURES, show=False)
    plt.tight_layout(); plt.savefig('ANN_SHAP_beeswarm.png', dpi=300); plt.close()
except Exception as e:
    with open('ANN_SHAP_NOTE.txt','w') as f:
        f.write(f"ANN SHAP skipped due to: {repr(e)}\n")

# -------------------------- 9) Concept Diagrams / Mimari Şemalar --------------
step("Create architecture diagrams (RF/ANN/Stacking)",
     "Mimari şemaları oluştur (RF/ANN/Stacking)")
graph_attr = {'rankdir': 'TB', 'splines': 'ortho', 'nodesep': '0.8'}
node_attr  = {'style': 'filled', 'shape': 'box', 'align': 'center', 'fontsize': '10', 'fontname': 'Helvetica'}
edge_attr  = {'fontsize': '9', 'fontname': 'Helvetica'}

rf_dot = Digraph('RandomForestConcept', graph_attr=graph_attr, node_attr=node_attr, edge_attr=edge_attr); rf_dot.attr(label='Random Forest Concept')
with rf_dot.subgraph(name='cluster_0') as c: c.attr(color='white'); c.node('Input', 'Full Training Data', shape='parallelogram', fillcolor='lightgrey')
with rf_dot.subgraph(name='cluster_1') as c: c.attr(label='Bootstrap & Feature Subspacing', style='dashed'); c.node('D1', 'Subset 1'); c.node('D2', 'Subset 2'); c.node('DN', 'Subset N')
with rf_dot.subgraph(name='cluster_2') as c: c.attr(label='Independent Trees', style='dashed'); c.node('T1', 'Tree 1', fillcolor='lightblue'); c.node('T2', 'Tree 2', fillcolor='lightblue'); c.node('TN', 'Tree N', fillcolor='lightblue')
rf_dot.node('AVG', 'Averaging', shape='ellipse', fillcolor='orange'); rf_dot.node('OUT', 'Final Prediction', shape='parallelogram', fillcolor='gold')
rf_dot.edge('Input','D1'); rf_dot.edge('Input','D2'); rf_dot.edge('Input','DN'); rf_dot.edge('D1','T1'); rf_dot.edge('D2','T2'); rf_dot.edge('DN','TN'); rf_dot.edge('T1','AVG'); rf_dot.edge('T2','AVG'); rf_dot.edge('TN','AVG'); rf_dot.edge('AVG','OUT')
rf_dot.render('random_forest_concept_FIXED', format='png', view=False)

ann_dot = Digraph('ANN_Architecture', graph_attr={'rankdir':'LR','splines':'line'}, node_attr=node_attr, edge_attr=edge_attr); ann_dot.attr(label='Artificial Neural Network (ANN) Architecture')
with ann_dot.subgraph(name='cluster_0') as c: c.attr(label='Input Layer'); c.attr(color='white'); c.node_attr.update(shape='parallelogram', fillcolor='lightgrey'); c.node('xn', f'Input Features ({len(TOP_FEATURES)})')
with ann_dot.subgraph(name='cluster_1') as c: c.attr(label='Hidden Layer 1'); c.attr(color='white'); c.node_attr.update(shape='circle', fillcolor='lightblue'); c.node('h1_1','Neuron'); c.node('h1_n','...')
with ann_dot.subgraph(name='cluster_2') as c: c.attr(label='Hidden Layer 2'); c.attr(color='white'); c.node_attr.update(shape='circle', fillcolor='lightblue'); c.node('h2_1','Neuron'); c.node('h2_n','...')
with ann_dot.subgraph(name='cluster_3') as c: c.attr(label='Output Layer'); c.attr(color='white'); c.node_attr.update(shape='parallelogram', fillcolor='gold'); c.node('y','Precipitation Forecast')
ann_dot.edge('xn','h1_1'); ann_dot.edge('xn','h1_n'); ann_dot.edge('h1_1','h2_1'); ann_dot.edge('h1_n','h2_n'); ann_dot.edge('h2_1','y'); ann_dot.edge('h2_n','y')
ann_dot.render('ann_architecture_FIXED', format='png', view=False)

stack_dot = Digraph('Stacking_Architecture', graph_attr=graph_attr, node_attr=node_attr, edge_attr=edge_attr); stack_dot.attr(label='Stacking Ensemble Architecture')
stack_dot.node('INP', f'Input Features ({len(TOP_FEATURES)})', shape='parallelogram', fillcolor='lightgrey')
with stack_dot.subgraph(name='cluster_base') as c: c.attr(label='Base Models', style='dashed'); c.node('XGB','XGBoost'); c.node('RF','RandomForest'); c.node('LGBM','LightGBM')
stack_dot.node('PRED','Base Predictions', shape='parallelogram', fillcolor='orange')
with stack_dot.subgraph(name='cluster_meta') as c: c.attr(label='Meta-Model', style='dashed'); c.node('META','Linear Regression')
stack_dot.node('OUT','Final Prediction', shape='parallelogram', fillcolor='gold')
stack_dot.edge('INP','XGB'); stack_dot.edge('INP','RF'); stack_dot.edge('INP','LGBM'); stack_dot.edge('XGB','PRED'); stack_dot.edge('RF','PRED'); stack_dot.edge('LGBM','PRED'); stack_dot.edge('PRED','META'); stack_dot.edge('META','OUT')
stack_dot.render('stacking_architecture_FIXED', format='png', view=False)

# -------------------------- 10) Zip & Download / Sıkıştır & İndir ------------
step("Zip all outputs and offer for download",
     "Tüm çıktıları ZIP’le ve indirmeye sun")
out_files = [
    'design_rainfall_table_with_CI.csv',
    'GUMBEL_return_levels_CI.png',
    'FINAL_TREND_ANALYSIS_fixed.png',
    'feature_importance_FIXED.png',
    'ULTIMATE_FINAL_comparison_FIXED.csv',
    'model_performance_barchart_FIXED.png',
    'ULTIMATE_FINAL_comparison_FIXED.png',
    'champion_model_scatter_FIXED.png',
    'EVENT_METRICS_top_models.csv',
    'RF_SHAP_bar.png','RF_SHAP_beeswarm.png',
    'LGBM_SHAP_bar.png','LGBM_SHAP_beeswarm.png',
    'XGB_SHAP_bar.png','XGB_SHAP_beeswarm.png',
    'random_forest_concept_FIXED.png',
    'ann_architecture_FIXED.png',
    'stacking_architecture_FIXED.png'
]
for f in ['ANN_SHAP_bar.png','ANN_SHAP_beeswarm.png','ANN_SHAP_NOTE.txt']:
    if os.path.exists(f): out_files.append(f)

zip_bytes = io.BytesIO()
with zipfile.ZipFile(zip_bytes, 'w', zipfile.ZIP_DEFLATED) as zf:
    for fn in out_files:
        if os.path.exists(fn):
            zf.write(fn)
zip_bytes.seek(0)
with open('NEWSTUDY_RESULTS_FIXED.zip','wb') as f:
    f.write(zip_bytes.read())

print("\n>>> Downloading ZIP / ZIP indiriliyor...")
files.download('NEWSTUDY_RESULTS_FIXED.zip')

print("\n=== ALL DONE / TÜM İŞLEMLER TAMAMLANDI ===")
```
